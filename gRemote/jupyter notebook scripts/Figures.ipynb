```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = "Linux Libertine O"

titleFont = {'family' : 'Linux Libertine O',
    'weight' : 'normal',
    'size'   : 19
    }

xyticksFont = {
    'family' : 'Linux Libertine O',
    'weight' : 'normal',
    'size'   : 16
}

labelFont = {
    'family' : 'Linux Libertine O',
    'weight' : 'normal',
    'size'   : 17
}

color1 = "#000000" #black
color2 = "r"       #red
color3 = "#B0B0B0" #gray
color4 = "b"       #blue
color5 = "#595959"
    
program_names = ['blender','dinoshade','scube','spot','stars','surfgrid','glxgears','drawelements']
```

```python
fig,axarr = plt.subplots(2,4)
fig.set_figheight(8)
fig.set_figwidth(18)

color_2client = "#EB2C30"
color_4client = "#0050A0"
color_6client = '#EE3B3B'
color_8client = "#202020"
color_10client = "#009F4D"
color_12client = "#991B5A"

for (ax,name) in zip(axarr.flat,program_names):
    ax.axhline(y=60,label='',color='#999999',linestyle='--',linewidth=2)
    ax.axhline(y=80,color='#999999',label='',linestyle='--',linewidth=2)
    data = pd.read_excel('all fps.xlsx',name).T
    print(ax)
    X = np.arange(1,21,1)
#     ms = np.hstack((np.arange(0,20,3),np.array(19)))
    ms = np.arange(0,20,2)
    mask = data==0
    data[mask] = 40
    ax.plot(X,data.loc[2][0:20],label="2 clients",color=color_2client,marker='s',markersize=5,markerfacecolor='none',markevery=ms.tolist(),linestyle='-')
    ax.plot(X,data.loc[4][0:20],label="4 clients",color=color_4client,marker='o',markersize=5,markerfacecolor='none',markevery=ms.tolist(),linestyle='--')
    ax.plot(X,data.loc[8][0:20],label="8 clients",color=color_8client,marker='<',markersize=5,markerfacecolor='none',markevery=ms.tolist(),linestyle='-')
    ax.plot(X,data.loc[10][0:20],label="",color=color_10client,marker='+',markerfacecolor='none',markevery=ms.tolist(),linestyle='--')
    ax.plot(X,data.loc[10][0:20],label="10 clients",color=color_10client,marker='x',markerfacecolor='none',markevery=ms.tolist(),linestyle='--')
    ax.plot(X,data.loc[6][0:20],label="12 clients",color=color_12client,marker='D',markersize=4,markerfacecolor='none',markevery=ms.tolist())
    ax.plot(X,data.loc[12][0:20],label="",color=color_12client,linestyle='-.')
#     data12 = data.loc[12][0:20]
#     for i in range(19):
#         if data12[i] == 40:
#             ax.scatter(x=i+1,y=data.loc[6][i],marker='D',s = 16,facecolors='none',color=color_12client,edgecolors=color_12client,label='')
#             ax.scatter(x=i,y=data.loc[6][i-1],marker='D',s= 16,facecolors='none',color=color_12client,edgecolors=color_12client,label='')
#             ax.scatter(x=i+2,y=data.loc[6][i+1],marker='D',s= 16,facecolors='none',color=color_12client,edgecolors=color_12client,label='')
    
    ax.set_yticks([40,50,60,70,80,90])
    ax.set_yticklabels([0,50,60,70,80,90],xyticksFont)
    ax.set_xticks(np.arange(0,22,5))
    ax.set_xticklabels(np.arange(0,22,5),xyticksFont)
    ax.set_title(name.capitalize() + '',titleFont)
    ax.set_ylabel('FPS',labelFont)
    ax.set_xlabel('Time(s)',labelFont)
    ax.spines["right"].set_visible(False)
    ax.spines["top"].set_visible(False)

    
axarr[0][0].legend(bbox_to_anchor=(0., 1.3, 4.75, .12),frameon=True,fontsize=18, loc='lower left',ncol=3, mode="expand", borderaxespad=0.)   
plt.tight_layout()
plt.savefig("/home/yinlan/Desktop/fig11_gremote_fps_line.pdf")
```
